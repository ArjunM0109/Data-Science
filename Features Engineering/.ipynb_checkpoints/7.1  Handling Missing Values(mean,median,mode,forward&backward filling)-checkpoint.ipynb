# 1. MEAN

### Mean: Provides a measure of the average value but can be skewed by outliers. Useful for normally distributed data

```python
si = SimpleImputer(strategy='mean')
```

```python
si.fit_transform(dataset[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']])
```

# 2. MEDIAN

### Median: Offers a middle ground that is not affected by outliers, making it suitable for skewed data.

```python
si_1 = SimpleImputer(strategy='median')
```

```python
si_1.fit_transform(dataset[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']])
```

# 3. MODE

### The mode of a dataset is the value or values that appear most frequently.

```python
n_data=dataset['Gender'].mode()[0]
```

```python
dataset['Gender'].fillna(n_data)
```

```python
dataset.select_dtypes(include='object').isnull().sum()  ,dataset.select_dtypes(include='object').columns
```

```python
for i in dataset.select_dtypes(include='object').columns:
    dataset[i].fillna(dataset[i].mode()[0],inplace=True)
```

```python
dataset.select_dtypes(include='object').isnull().sum()
```
