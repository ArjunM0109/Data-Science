# STANDARDIZATION

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r'C:\Users\acer\Desktop\Python\Data Science\Features Engineering\data\loan.csv')
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset['ApplicantIncome'])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
ss.fit(dataset[['ApplicantIncome']])
```

```python
dataset['ApplicantIncome_ss']=pd.DataFrame(ss.transform(dataset[['ApplicantIncome']]),columns=['x'])
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.distplot(dataset['ApplicantIncome_ss'])

plt.subplot(1,2,2)
sns.distplot(dataset['ApplicantIncome'])
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(dataset['ApplicantIncome_ss'])

plt.subplot(1,2,2)
sns.histplot(dataset['ApplicantIncome'])
```

# NORMALIZATION

```python
data_set = pd.read_csv(r'C:\Users\acer\Desktop\Python\CSV FILES\loan.csv')
data_set.head(3)
```

```python
data_set.describe()
```

```python
sns.distplot(data_set['CoapplicantIncome'])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
```

```python
ms.fit(data_set[['CoapplicantIncome']])

# Transform the 'CoapplicantIncome' column
data_set['CoapplicantIncome_ms'] = ms.transform(data_set[['CoapplicantIncome']])
```

```python
data_set.head(3)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title('before')
sns.distplot(data_set['CoapplicantIncome'])

plt.subplot(1,2,2)
plt.title('after')
sns.distplot(data_set['CoapplicantIncome_ms'])
plt.show()
```
