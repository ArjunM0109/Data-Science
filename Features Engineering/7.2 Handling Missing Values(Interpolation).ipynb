# 4. HOT DUCK Imputation

```python
# Hot Deck Imputation
def hot_deck_impute(series):
    """Randomly selects a non-missing value to fill NaNs."""
    non_na_values = series.dropna().values
    return series.apply(lambda x: np.random.choice(non_na_values) if pd.isna(x) else x)
```

```python
# Apply hot deck imputation to the 'Sales' column and create a new column for the imputed values
df['Sales_Hot_Deck_Imputed'] = hot_deck_impute(df['Sales'])
```
